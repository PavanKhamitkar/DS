# Filter data by species

```python
species_data = {
    'Iris-setosa': df[df['species'] == 'Iris-setosa'],
    'Iris-versicolor': df[df['species'] == 'Iris-versicolor'],
    'Iris-virginica': df[df['species'] == 'Iris-virginica']
}
```

```python
print(species_data)
```

# Calculate statistics (assuming numeric columns for petal length, petal width, etc.)

```python
species_stats = {}
for species, data in species_data.items():
    species_stats[species] = data[['petal_length', 'petal_width', 'sepal_length', 'sepal_width']].describe()
```

# Display results

```python
for species, stats in species_stats.items():
    print(f"\n**{species} Statistics:")
    print(stats)
```
